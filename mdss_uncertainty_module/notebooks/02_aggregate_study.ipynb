# Aggregate Study

Analyze batch processing results.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load results
df = pd.read_csv('../data/results/results.csv')
print(f"Loaded {len(df)} records")
df.head()
```

```python
# Summary
print("Risk distribution:")
print(df['risk_level'].value_counts())

print(f"\nHuman review rate: {df['requires_review'].mean():.1%}")
print(f"Mean confidence: {df['confidence'].mean():.3f}")
```

```python
# Visualize
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

df['risk_level'].value_counts().plot(kind='bar', ax=axes[0], color=['green', 'yellow', 'orange', 'red'])
axes[0].set_title('Risk Distribution')

axes[1].hist(df['confidence'], bins=50)
axes[1].set_title('Confidence Distribution')
axes[1].set_xlabel('Confidence Score')

plt.tight_layout()
plt.show()
```
